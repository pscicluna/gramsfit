```python
tDLDD = Table.read('M1.00.DL.DD.stb', format = 'ascii')
tOcDD = Table.read('M1.00.Oc.DD.stb', format = 'ascii')
tOwDD = Table.read('M1.00.Ow.DD.stb', format = 'ascii')
tDL = Table.read('M1.00.DL.stb', format = 'ascii')
tOc = Table.read('M1.00.Oc.stb', format = 'ascii')
tOw = Table.read('M1.00.Ow.stb', format = 'ascii')
plt.plot(tDLDD['lambda'], tDLDD['fInp'], '--', label = 'Photosphere')
plt.plot(tDLDD['lambda'], tDLDD['fTot'], 'r', label='DL DD')
plt.plot(tOcDD['lambda'], tOcDD['fTot'], 'b', label='Oc DD')
plt.plot(tOwDD['lambda'], tOwDD['fTot'], 'g', label='Ow DD')
plt.plot(tDL['lambda'], tDL['fTot'], 'r', label='DL', alpha=0.3)
plt.plot(tOc['lambda'], tOc['fTot'], 'b', label='Oc', alpha=0.3)
plt.plot(tOw['lambda'], tOw['fTot'], 'g', label='Ow', alpha=0.3)
# plt.xscale('log')
# plt.yscale('log')
plt.legend(loc='best')
plt.ylim(1e-3, 1)
plt.xlim(.1, 30)
```
